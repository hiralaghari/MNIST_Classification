## MNIST Classification Project

```python
## Load libraries and dataset
import tensorflow
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.datasets import mnist
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.metrics import confusion_matrix, classification_report
from tensorflow.keras.callbacks import TensorBoard
import time
```

```python
(train_images, train_labels), (test_images, test_labels) = keras.datasets.mnist.load_data()

# Normalize the images to a range of 0 to 1 by dividing by 255
train_images = train_images / 255.0
test_images = test_images / 255.0

# Reshape the data to fit the model (28x28 pixels, 1 channel)
train_images = train_images.reshape((train_images.shape[0], 28, 28, 1))
test_images = test_images.reshape((test_images.shape[0], 28, 28, 1))
```

#### The CNN Architecture
A Convolutional Neural Network (CNN) typically consists of three types of layers: **convolutional layers**, **pooling layers**, and **fully connected layers**. Each layer has a specific role in processing the input image data to ultimately classify it.

1. **Input Layer**

The input to the network is an image of size 28x28 pixels, with a single color channel (grayscale). This is why we specify the input shape as (28, 28, 1).
- Conv2D(32, (3, 3)): This layer applies 32 filters (each of size 3x3) to the input image. These filters scan the image and capture different features like edges, corners, and textures.
- Activation='relu': The Rectified Linear Unit (ReLU) activation function is applied to introduce non-linearity to the model, which helps the network learn more complex patterns.
- Input shape: (28, 28, 1) specifies the shape of the input image.
2. **Convolutional Layer**

Convolutional layers are the core building blocks of a CNN. They perform the convolution operation, which involves sliding the filters across the input image and computing the dot product between the filter and the local regions of the image. The result is a feature map that highlights the presence of certain features detected by the filters.
- Conv2D(64, (3, 3)): This layer applies 64 filters of size 3x3 to the output of the previous layer. Increasing the number of filters allows the network to learn more features.
3. **Pooling Layer**

Pooling layers are used to reduce the spatial dimensions (width and height) of the feature maps. This downsampling helps to reduce the computational load and prevent overfitting. The most common type of pooling is max pooling.
- MaxPooling2D((2, 2)): This layer reduces the size of the feature maps by taking the maximum value in each 2x2 block of the input feature map. It effectively downsamples the image while retaining the most important features.
4. **Flatten Layer**

After the convolutional and pooling layers, the feature maps are flattened into a single vector. This is necessary because the fully connected (dense) layers that follow require a 1D input.
- Flatten(): This layer converts the 2D output from the previous layer into a 1D vector.
5. **Dense Layers (Fully Connected Layers)**

These layers are the final part of the CNN, where the actual classification takes place. Dense layers connect every neuron in the layer to every neuron in the previous layer.
- Dense(64, activation='relu'): This layer has 64 neurons, each connected to every neuron in the flattened layer. The ReLU activation function is used to introduce non-linearity.
- Dense(10, activation='softmax'): The final dense layer has 10 neurons, corresponding to the 10 digit classes (0-9). The softmax activation function is used here to output a probability distribution over the 10 classes, with the sum of all probabilities equal to 1.

```python
model = keras.models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

model.summary()  # To see the structure of the model
```

#### Model Compilation
1. **Optimizer**

- **What It Is**: The optimizer is an algorithm or method used to update the weights of the neural network to minimize the loss during training. It determines how the model learns from the data by adjusting the weights.
- **Why It's Important**: The choice of optimizer affects how quickly and effectively the model converges to a solution.
- Adam Optimizer:
    - Adaptive Moment Estimation (Adam) is a widely used optimizer that combines the benefits of two other optimizers: AdaGrad and RMSProp.
    - Adaptive Learning Rates: Adam adjusts the learning rate for each parameter individually, using estimates of first (mean) and second (uncentered variance) moments of the gradients. This makes it more effective in handling sparse gradients and varying learning rates.
    - Momentum: Adam incorporates momentum, which helps accelerate gradient vectors in the right direction, leading to faster converging.
2. **Loss Function**

- **What It Is**: The loss function measures how well the model's predictions match the actual target values. It's a mathematical function that the optimizer tries to minimize during training.
- **Why It's Important**: The loss function directly influences the training process by guiding the optimizer in adjusting the model's weights.
- sparse_categorical_crossentropy Loss:
    - **Categorical Crossentropy**: This loss function is commonly used for classification tasks where the model's output is a probability distribution over multiple classes. It measures the difference between the predicted probability distribution and the true distribution (which is 1 for the correct class and 0 for others).
    - **Sparse**: The term "sparse" indicates that the labels are provided as integers (e.g., 0, 1, 2,...) rather than one-hot encoded vectors. This makes it efficient for multi-class classification problems, especially when dealing with a large number of classes.
3. **Metrics**

- **What They Are**: Metrics are functions used to evaluate the performance of the model. They are not used by the optimizer to train the model but provide insight into how well the model is doing.
- **Why They're Important**: Metrics help monitor the model's performance during training and evaluation, providing a way to understand how well the model is generalizing to new data.
- accuracy Metric:
   - Accuracy measures the proportion of correct predictions made by the model out of all predictions.
    - **Why Accuracy**: For classification tasks, especially those with balanced classes, accuracy is a straightforward and interpretable metric to track how well the model is performing.

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

#### Choosing Epochs & Batch Sizes, and monitoring performance

1. **Batch Size**: Batch size is the number of training examples utilized in one iteration (or forward/backward pass).
- **Small Batch Size (e.g., 32, 64):**
    - **Pros**: More frequent weight updates, which can lead to a more robust learning process.
    - **Cons**: Takes longer to complete an epoch, and the training might be noisier (more fluctuations in the loss curve).
- **Large Batch Size (e.g., 256, 512):**
    - **Pros**: Faster computation as it makes better use of the hardware (e.g., GPUs).
    - **Cons**: Fewer weight updates, which can make the learning process less stable and potentially miss out on finding the optimal solution.

2. **Epochs**: An epoch is one complete pass through the entire training dataset.
- How It Works:
    - During training, the model is updated after each batch. After processing all the batches (one epoch), the model starts the next epoch with updated weights.
    - If you set the number of epochs to 10, the model will see each training example 10 times (in different batches).
- Choosing the Number of Epochs:
    - **Too Few Epochs**: The model may not have enough time to learn the patterns in the data (underfitting).
    - **Too Many Epochs**: The model might learn too much, including noise and random fluctuations (overfitting).
    - **Rule of Thumb**: Start with a reasonable number (e.g., 10-20 epochs), and monitor the performance on the validation set. If validation accuracy continues to improve, consider increasing the number of epochs. If validation accuracy starts to plateau or decrease, it might be time to stop (or use early stopping).

```python
model_1 = model.fit(train_images, train_labels, epochs=5, batch_size=64, validation_data=(test_images, test_labels))
```

```python

# Plot training & validation accuracy values
plt.plot(model_1.history['accuracy'])
plt.plot(model_1.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()

# Plot training & validation loss values
plt.plot(model_1.history['loss'])
plt.plot(model_1.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()
```

#### Early Stopping
- **What It Is**: Early stopping is a technique to stop training when the model’s performance on the validation data starts to degrade (i.e., when it starts to overfit).
- **Parameters**:
    - **Monitor**: The metric you want to monitor (e.g., val_loss).
    - **Patience**: Number of epochs to wait after the last improvement before stopping.
    - **Restore Best Weights**: If True, it restores the model to the state from the epoch with the best validation loss.

```python
early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

model_2 = model.fit(train_images, train_labels, epochs=50, batch_size=64, validation_data=(test_images, test_labels), callbacks=[early_stopping])
```

```python
# Plot training & validation accuracy values
plt.plot(model_2.history['accuracy'])
plt.plot(model_2.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()

# Plot training & validation loss values
plt.plot(model_2.history['loss'])
plt.plot(model_2.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()
```

```python

# Predict the classes for test data
predictions = model.predict(test_images)
predicted_classes = np.argmax(predictions, axis=1)

# Generate confusion matrix
cm = confusion_matrix(test_labels, predicted_classes)
print(cm)

# Generate classification report
report = classification_report(test_labels, predicted_classes)
print(report)
```
